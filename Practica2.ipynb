```python
df_weather = pd.read_csv('/content/drive/MyDrive/Ciència de les dades/Tipologia i cicle de vida de les dades/Pràctica 2/daily_weather_2020.csv')
df_weather.head()
```

```python
df_weather.columns = ['Country name' if x=='Country/Region' else x for x in df_weather.columns]
df_weather.head()
```

```python
weather_by_country = df_weather.groupby('Country name').mean()
weather_by_country.head()
```

```python
weather_by_country.columns
```

```python
interesting_columns = ['temperatureHigh', 'temperatureLow', 'windSpeed', 'precipAccumulation']
weather_by_country_sub = weather_by_country[interesting_columns]
weather_by_country_sub.head()
```

```python
weather_by_country_sub['temperatureAvg'] = (weather_by_country_sub['temperatureHigh'] + weather_by_country_sub['temperatureLow'])/2
weather_by_country_sub.head()
```

```python
df_combined_data = df2021.merge(weather_by_country_sub, on='Country name')
df_combined_data.head()
```

```python
df_combined_data.replace("", np.nan, inplace = True)
```

```python
df_combined_data.isna().sum()
```

```python
null_rows = df_combined_data[df_combined_data['precipAccumulation'].isnull()]
null_rows
```

```python
means = df_combined_data.groupby('Regional indicator')['precipAccumulation'].transform(np.mean)
df_combined_data['precipAccumulation'] = df_combined_data['precipAccumulation'].fillna(means)
```
